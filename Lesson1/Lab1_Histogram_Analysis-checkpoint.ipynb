# Ćwiczenie 1: Analiza histogramu obrazów
Celem ćwiczenia jest zrozumienie roli histogramu obrazu oraz wpływu różnych transformacji histogramu na jakość obrazów.

## Podstawy teoretyczne
- Histogram obrazu definiowany jest jako rozkład liczby pikseli względem ich wartości intensywności.
- Transformacje histogramu obejmują:
  1. Wyrównywanie histogramu
  2. Rozciąganie histogramu
  3. Skalowanie
  4. Wygładzanie i wyostrzanie obrazu

- Do oceny jakości stosuje się metryki **PSNR** i **SSIM**:
$$PSNR = 10 \log_{10}\left(\frac{MAX^2}{MSE}\right)$$
$$SSIM(x,y) = \frac{(2\mu_x \mu_y + c_1)(2\sigma_{xy}+c_2)}{(\mu_x^2+\mu_y^2+c_1)(\sigma_x^2+\sigma_y^2+c_2)}$$

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.metrics import peak_signal_noise_ratio as psnr, structural_similarity as ssim

# --- Funkcje pomocnicze ---

def load_image(path):
    """Wczytanie obrazu w skali szarości"""
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise FileNotFoundError(f"Nie znaleziono pliku: {path}")
    return img

def show_image_and_hist(img, title="Obraz"):
    """Wyświetlenie obrazu i jego histogramu"""
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))
    
    # obraz
    ax[0].imshow(img, cmap='gray')
    ax[0].set_title(title)
    ax[0].axis("off")
    
    # histogram
    ax[1].hist(img.ravel(), bins=256, range=(0, 256), color='black')
    ax[1].set_title("Histogram")
    ax[1].set_xlim([0, 255])
    
    plt.tight_layout()
    plt.show()

# --- Funkcje transformacji ---

def hist_equalization(img):
    return cv2.equalizeHist(img)

def hist_stretching(img):
    return cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX)

def hist_scaling(img, alpha=1.2):
    return cv2.convertScaleAbs(img, alpha=alpha, beta=0)

def smoothing(img):
    return cv2.GaussianBlur(img, (5, 5), 0)

def sharpening(img):
    kernel = np.array([[0, -1, 0],
                       [-1, 5, -1],
                       [0, -1, 0]])
    return cv2.filter2D(img, -1, kernel)

# --- Przykład użycia ---
# Wczytaj obraz testowy (np. lena.png)
img = load_image("pink.png")
show_image_and_hist(img, "Oryginalny obraz")

eq = hist_equalization(img)
show_image_and_hist(eq, "Wyrównany histogram")

st = hist_stretching(img)
show_image_and_hist(st, "Rozciąganie histogramu")

sc = hist_scaling(img, alpha=1.5)
show_image_and_hist(sc, "Skalowanie histogramu")

sm = smoothing(img)
show_image_and_hist(sm, "Wygładzanie")

sh = sharpening(img)
show_image_and_hist(sh, "Wyostrzanie")

# --- Ocena jakości ---
print("Ocena jakości (PSNR, SSIM względem oryginału):")
for name, transformed in [("Equalizacja", eq), ("Rozciąganie", st),
                          ("Skalowanie", sc), ("Wygładzanie", sm), ("Wyostrzanie", sh)]:
    print(f"{name}: PSNR={psnr(img, transformed):.2f}, SSIM={ssim(img, transformed):.4f}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Wczytanie obrazu (OpenCV domyślnie wczytuje w BGR)
img_bgr = cv2.imread("pink.png")
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Rozdzielenie na kanały
R, G, B = cv2.split(img_rgb)

# Wizualizacja
plt.figure(figsize=(12,8))

# Oryginalny obraz
plt.subplot(2,2,1)
plt.imshow(img_rgb)
plt.title("Obraz RGB")
plt.axis("off")

# Kanał R
plt.subplot(2,2,2)
plt.imshow(R, cmap='Reds')
plt.title("Kanał R (Red)")
plt.axis("off")

# Kanał G
plt.subplot(2,2,3)
plt.imshow(G, cmap='Greens')
plt.title("Kanał G (Green)")
plt.axis("off")

# Kanał B
plt.subplot(2,2,4)
plt.imshow(B, cmap='Blues')
plt.title("Kanał B (Blue)")
plt.axis("off")

plt.tight_layout()
plt.show()

# Wyświetlenie macierzy intensywności dla przykładowego kanału (np. R)
print("Macierz intensywności kanału R:")
print(R)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Wczytanie obrazu
img_bgr = cv2.imread("pink.png")
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Rozdzielenie kanałów
R, G, B = cv2.split(img_rgb)

# Funkcja do wizualizacji 3D macierzy intensywności
def plot_channel_3d(channel, title, cmap):
    h, w = channel.shape
    X, Y = np.meshgrid(np.arange(w), np.arange(h))
    Z = channel

    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111, projection="3d")
    surf = ax.plot_surface(X, Y, Z, cmap=cmap, linewidth=0, antialiased=False)
    ax.set_title(title)
    ax.set_xlabel("X (szerokość)")
    ax.set_ylabel("Y (wysokość)")
    ax.set_zlabel("Intensywność [0-255]")
    fig.colorbar(surf, shrink=0.5, aspect=10)
    plt.show()

# --- 2D podgląd obrazu RGB i kanałów ---
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(img_rgb)
plt.title("Obraz RGB")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(R, cmap="Reds")
plt.title("Kanał R")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(G, cmap="Greens")
plt.title("Kanał G")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(B, cmap="Blues")
plt.title("Kanał B")
plt.axis("off")

plt.tight_layout()
plt.show()

# --- 3D wizualizacje macierzy intensywności ---
plot_channel_3d(R, "3D wizualizacja kanału R (Red)", "Reds")
plot_channel_3d(G, "3D wizualizacja kanału G (Green)", "Greens")
plot_channel_3d(B, "3D wizualizacja kanału B (Blue)", "Blues")
```
